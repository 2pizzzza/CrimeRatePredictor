```python
def convert_time(time_value):
    if pd.isna(time_value):
        return np.nan, np.nan
    
    time_str = str(int(time_value)).zfill(4)
    hours = int(time_str[:2]) if len(time_str) >= 2 else 0
    minutes = int(time_str[2:]) if len(time_str) > 2 else 0
    return hours, minutes
```

```python
def create_crime_map(data, num_points=1000):
    sample_data = data.sample(min(num_points, len(data)))
    
    avg_lat = sample_data['LAT'].mean()
    avg_lon = sample_data['LON'].mean()
    crime_map = folium.Map(location=[avg_lat, avg_lon], zoom_start=11)
    
    heat_data = [[row['LAT'], row['LON']] for _, row in sample_data.iterrows()]
    HeatMap(heat_data).add_to(crime_map)
    
    return crime_map

crime_map = create_crime_map(df)
```

```python
def predict_future_crime(model, last_data, encoder, n_periods=6):
    """
    Прогнозирует уровень преступности на n_periods месяцев вперед
    
    Args:
        model: Обученная модель
        last_data: Последние данные для прогнозирования
        encoder: OneHotEncoder для районов
        n_periods: Количество месяцев для прогноза
    
    Returns:
        DataFrame с прогнозами
    """
    # Создаем копию последних данных
    future_data = last_data.copy()
    predictions = []
    
    # Получаем уникальные районы
    areas = future_data['AREA NAME'].unique()
    
    for area in areas:
        area_data = future_data[future_data['AREA NAME'] == area].sort_values('date').iloc[-1:].copy()
        
        # Прогнозируем на n_periods вперед
        for i in range(n_periods):
            # Подготавливаем данные для следующего месяца
            last_date = area_data['date'].iloc[-1]
            next_month = last_date + pd.DateOffset(months=1)
            
            next_data = area_data.iloc[-1:].copy()
            next_data['date'] = next_month
            next_data['Year'] = next_month.year
            next_data['Month'] = next_month.month
            
            # Обновляем сезонные признаки
            next_data['month_sin'] = np.sin(2 * np.pi * next_data['Month'] / 12)
            next_data['month_cos'] = np.cos(2 * np.pi * next_data['Month'] / 12)
            
            # Получаем признаки
            X_next = next_data[features].values
            area_encoded = encoder.transform(next_data[['AREA NAME']])
            X_next = np.hstack([X_next, area_encoded])
            
            # Прогнозируем
            y_pred = model.predict(X_next)[0]
            next_data['crime_count'] = max(0, y_pred)  # Не может быть отрицательным
            
            # Обновляем скользящие средние
            next_data['prev_month_crimes'] = area_data['crime_count'].iloc[-1]
            next_data['prev_2_months_avg'] = (area_data['crime_count'].iloc[-1] + 
                                              area_data['prev_month_crimes'].iloc[-1]) / 2
            next_data['prev_3_months_avg'] = (area_data['crime_count'].iloc[-1] + 
                                             area_data['prev_month_crimes'].iloc[-1] + 
                                             area_data['prev_2_months_avg'].iloc[-1]) / 3
            
            # Добавляем в результаты
            predictions.append(next_data)
            area_data = pd.concat([area_data, next_data])
    
    # Объединяем все прогнозы
    future_predictions = pd.concat(predictions)
    return future_predictions
```
